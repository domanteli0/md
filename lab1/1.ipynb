```python
from sympy import Function, dsolve, Derivative, E, Eq, lambdify, symbols
from sympy import print_python, print_maple_code, print_latex, latex
from sympy.abc import x, y
import sympy as sm
from pipe import Pipe, map as m
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt


# pdisplay = Pipe(display)
# simplify = Pipe(lambda r: r.simplify())
```

```python
# Albegraic
y = Function('y')
# eq = (Derivative(y(x), x, x) + 9*y(x), y(x))

# eq = [x**2 + y(x)**2, y(x)]
eq = Eq(y(x) * (E**x + 1) - E**x * Derivative(y(x), x), 0)

# for result in dsolve(*eq):
#   result | simplify | pdisplay
solution = dsolve(eq)
print_maple_code(solution)

C = symbols('C')
solution = solution.subs('C1', C)
```

```python

# from helpers import display_sol
sol = solution.rhs

# Convert the solution into a numerical function using lambdify for different C1 values
sol1 = lambdify(x, sol.subs('C', 1))
sol2 = lambdify(x, sol.subs('C', 2))
sol3 = lambdify(x, sol.subs('C', 0.5))

x_vals = np.linspace(-1, 5, 400)
y_vals1 = sol1(x_vals)
y_vals2 = sol2(x_vals)
y_vals3 = sol3(x_vals)

plt.figure(figsize=(6, 6))

plt.plot(x_vals, y_vals1, label='C = 1', color='blue')
plt.plot(x_vals, y_vals2, label='C = 2', color='green')
plt.plot(x_vals, y_vals3, label='C = 0.5', color='red')

plt.title(f"${latex(solution)}$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
